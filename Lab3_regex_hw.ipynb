```python
import io
text = io.open("Spotify Charts.html", mode="r", encoding="utf-8").read()
text
```

```python
import re
```

```python
x = re.findall("http.?://[^\"' >]+",text)
y = re.findall("#[0-9a-fA-F]{3,6}",text)
z = re.findall("by ([\w ]+[^<]+)",text)
```

```python
import json
```

```python
jsonString = json.dumps(x)
print(jsonString)
```

```python
with open('urls.json', 'w') as json_file:
    json.dump(x, json_file)
```

```python
with open('hex.json', 'w') as json_file:
    json.dump(y, json_file)
```

```python
with open('artist.json', 'w') as json_file:
    json.dump(z, json_file)
```
